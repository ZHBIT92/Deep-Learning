```python
import re  
import sys  
import urllib  
import os  
  
import requests
```

```python
def down_pic(pic_urls,localPath):  
    if not os.path.exists(localPath):  # 新建文件夹  
        os.mkdir(localPath)  
    """给出图片链接列表, 下载图片"""  
    j = 44*20
    for i, pic_url in enumerate(pic_urls):  
        try:  
            pic = requests.get(pic_url, timeout=20)  
            string = str(i+1) + '.jpg'  
            with open(localPath + '%d.jpg' % int(i+j), 'wb')as f:  
                f.write(pic.content)  
          #  with open(string, 'wb') as f:  
          #      f.write(pic.content)  
                print('成功下载第%s张图片: %s' % (str(i+ 1), str(pic_url)))  
        except Exception as e:  
            print('下载第%s张图片时失败: %s' % (str(i+ 1), str(pic_url)))  
            print(e)  
            continue  
```

```python
def get_onepage_urls(onepageurl):  
    if not onepageurl:  
        print('执行结束')  
        return [], ''  
    try:  
        html = requests.get(onepageurl).text  
    except Exception as e:  
        print(e)  
        return pic_urls 
    pic_urls = re.findall('"objURL":"(.*?)",', html, re.S)  
    return pic_urls
```

```python
keyword = '星野源'  # 关键词, 改为你想输入的词即可  
page = 50   # 爬取页数,一页20张图片
url_init_first = r'http://image.baidu.com/search/flip?tn=baiduimage&ie=utf-8&word='
url_init = url_init_first + urllib.parse.quote(keyword, safe='/')  

all_url_page = []  
all_pic_urls = []
for i in range(40,page):
    url_page = url_init+'&pn='+str(i*20)+'&gsm=64&ct=&ic=0&lm=-1&width=0&height=0' 
    onepage_urls = get_onepage_urls(url_page)
    all_pic_urls.extend(onepage_urls)

down_pic(list(set(all_pic_urls)),'image/other/')# 保存位置也可以修改  
```
