```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
print ("Train data shape:", train.shape)
print ("Test data shape:", test.shape)
```

因为测试数据不包括最终销售价格信息，所以 *test* 比 *train* 少一列

```python
train.head()
```

* SalePrice - 以美元为单位的物业销售价格 这是你试图预测的目标变量
* MSSubClass - 建筑类
* MSZoning - 一般的分区划分

```python
import matplotlib.pyplot as plt
plt.style.use(style='ggplot')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
train.SalePrice.describe()
```

Series.describe()函数获取销售价格的详细信息

```python
# 用plt.hist()绘制直方图SalePrice来检测偏移量
print ("Skew is:", train.SalePrice.skew())
plt.hist(train.SalePrice, color='blue')
plt.show()
```

np.log()会改变目标变量，使用np.log()转换train.SalePric和计算偏斜度

```python
target = np.log(train.SalePrice)
print ("Skew is:", target.skew())
plt.hist(target, color='blue')
plt.show()
```

可以看到数据更接近正态分布，说明已经改善了数据的偏移度

## 接下来我们检查数字特征
.select_dtypes()方法将返回匹配指定数据类型的列的子集

```python
numeric_features = train.select_dtypes(include=[np.number])
numeric_features.dtypes
```

DataFrame.corr()方法显示列之间的相关性（或关系）

```python
corr = numeric_features.corr()

print (corr['SalePrice'].sort_values(ascending=False)[:5], '\n')
print (corr['SalePrice'].sort_values(ascending=False)[-5:],'\n')
```

前五个特征与最正相关SalePrice，而后五个特征与负相关性最大。

```python
train.OverallQual.unique()
```

.unique()方法来去除重复的元素以获得唯一的值，发现OverallQual数据是整数值在区间1到10以下

### 对OverallQual和SalePrice之间的关系进行探究
pivot_table()建立一个数据透视表，np.median()返回数组元素的中位数

```python
quality_pivot = train.pivot_table(index='OverallQual',
                                  values='SalePrice', aggfunc=np.median)
quality_pivot
```

使用Series.plot()方法创建一个条形图

```python
quality_pivot.plot(kind='bar', color='blue')
plt.xlabel('Overall Quality')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

可以发现随着整体质量的提高，中位销售价格严重上涨

### 对GrLivArea(地面生活区)、GarageArea(车库面积)和SalePrice之间的关系进行探究

```python
plt.scatter(x=train['GrLivArea'], y=target)
plt.ylabel('Sale Price')
plt.xlabel('Above grade (ground) living area square feet')
plt.show()
```

我们看到居住面积的增加与价格上涨相对应，但接下来我们看下GarageArea与SalePrice之间的关系

```python
plt.scatter(x=train['GarageArea'], y=target)
plt.ylabel('Sale Price')
plt.xlabel('Garage Area')
plt.show()
```

发现许多家庭的Garage Area值为0，说明他们没有车库。

接下来我们将创建一个删除了一些异常值的新数据框

```python
train = train[train['GarageArea'] < 1200]
plt.scatter(x=train['GarageArea'], y=np.log(train.SalePrice))
plt.xlim(-200,1600) # This forces the same scale as before
plt.ylabel('Sale Price')
plt.xlabel('Garage Area')
plt.show()
```
